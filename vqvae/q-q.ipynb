```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

codebook_1 = np.load("../results/original/original.npy")
codebook_2 = np.load("../results/gaussian/gaussian.npy")

codebook_1_flat = codebook_1.flatten()
codebook_2_flat = codebook_2.flatten()

plt.figure(figsize=(8, 6), dpi=200)
stats.probplot(codebook_2_flat, dist="norm", plot=plt)
plt.title("Q-Q Plot for Noisy Codebook (Gaussian) vs Theoretical Normal Distribution")
plt.show()
```

    Deviation from the Line:

    The data points (blue) do not perfectly align with the red diagonal line, especially at the tails (both extremes). This indicates that the distribution of your noisy codebook data deviates from the theoretical normal distribution, especially for very low and very high quantiles.
    S-Shaped Curve:

    The Q-Q plot shows an S-shaped curve, suggesting that the noisy codebook has heavier tails than a normal distribution. This means that there are more extreme values (outliers) in your noisy codebook than what would be expected under a normal distribution.
    Middle Quantiles Fit Better:

    In the middle quantile range (around 0), the points are closer to the line. This suggests that the central part of the distribution (closer to the mean) of your noisy codebook is somewhat normal, but the deviations increase in the tails.
    Inference:
    Non-Normal Distribution: The noisy codebook does not follow a perfect normal distribution, particularly due to the presence of outliers or heavy tails.
    Impact of Noise: The Gaussian noise seems to have introduced more extreme values (outliers) into the data, which is why the tail ends (both left and right) of the plot deviate more from the expected normal quantiles.
    Central Tendency: While the overall distribution deviates from normality, the central part of the distribution (quantiles near the mean) still behaves somewhat normally.
