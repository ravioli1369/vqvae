```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt

codebook_1 = np.load("../results/original/original.npy")
codebook_2 = np.load("../results/gaussian/gaussian.npy")

if codebook_1.size == 0 or codebook_2.size == 0:
    raise ValueError("One of the codebook files is empty.")

codebook_1_flat = codebook_1.flatten()
codebook_2_flat = codebook_2.flatten()

kde_1 = gaussian_kde(codebook_1_flat)
kde_2 = gaussian_kde(codebook_2_flat)

x_vals = np.linspace(
    np.min([codebook_1_flat, codebook_2_flat]),
    np.max([codebook_1_flat, codebook_2_flat]),
    1000,
)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, kde_1(x_vals), label="Codebook 1 KDE (clean data)", color="blue")
plt.plot(x_vals, kde_2(x_vals), label="Codebook 2 KDE (gaussian noise)", color="green")
plt.fill_between(x_vals, kde_1(x_vals), alpha=0.3, color="blue")
plt.fill_between(x_vals, kde_2(x_vals), alpha=0.3, color="green")

plt.title("Kernel Density Estimation (KDE) of Codebook Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

Peak Similarity:
Both KDEs exhibit a peak centered around 0, which indicates that the majority of the vectors in both codebooks (clean and noisy) are clustered near 0. This suggests that the Gaussian noise we added has not drastically shifted the central location of the distribution for the codebook vectors.

Variance/Spread:
The green KDE (Gaussian noise) is slightly narrower and taller than the blue KDE (clean data), indicating that the noisy data has less variance than the clean data. This could mean that the noise is not uniformly distributed across all vectors or dimensions but may be constraining the vectors toward the center more tightly.

Tail Behavior:
The tails of both distributions (far left and right of the peaks) show similar behavior, tapering off around the same points. This suggests that the outliers or extremes in both distributions are roughly equivalent, implying that the noise we introduced did not create significantly more extreme outliers.
