```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

codebook_1 = np.load("../results/original/original.npy")
codebook_2 = np.load("../results/poisson/poisson.npy")


cosine_sim = cosine_similarity(codebook_1, codebook_2)
plt.figure(figsize=(8, 6))
# sns.heatmap(cosine_sim, cmap="coolwarm", annot=False, origin='lower')
plt.imshow(cosine_sim, cmap="coolwarm", origin="lower")
plt.colorbar()
plt.title("Cosine Similarity between Codebook 1(Clean) and Codebook 2(Gaussian Noise)")
plt.show()
```

Cosine Similarity Distribution:

The values of cosine similarity range from approximately -0.75 to 0.75, as shown by the color bar on the right. Positive values (reds) indicate higher similarity between vectors, while negative values (blues) indicate dissimilarity.
There is a mix of red, blue, and neutral (whites) areas, meaning that some vectors from the clean codebook are quite similar to those in the noisy codebook, while others show significant dissimilarity.
Scattered Pattern:

There is no clear clustering of similarity across the heatmap, indicating that the noise has an inconsistent effect on the codebook vectors. Some vectors remain relatively similar after noise is introduced (reddish patches), while others become more dissimilar (blue patches).
Symmetry:

The heatmap does not show any obvious patterns of symmetry, which means the noise has affected different vectors in varying ways without any systematic bias toward certain vector pairs.
Regions of High and Low Similarity:

The red areas (values close to 0.75) show that for certain vector pairs, the clean and noisy codebooks remain relatively similar, despite the Gaussian noise.
The blue areas (values closer to -0.75) indicate strong dissimilarity between the vectors in the two codebooks, showing that the noise has shifted the representations significantly for certain parts of the dataset.
Inference:
Inconsistent Noise Effect: The Gaussian noise does not have a uniform effect across all vectors in the codebook. Some vectors are still quite similar, while others are significantly altered.
Vector-Specific Impact: The variability in cosine similarity suggests that some vectors are more robust to noise, while others are more sensitive. This could be important if certain features in the dataset are more noise-sensitive than others.
Diverse Changes in Representation: The overall lack of a clear structure or consistent pattern in the heatmap implies that the noise introduces diverse and non-systematic changes in the vector representations.
