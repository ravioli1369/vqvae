```python
og_codebooks, speckle_codebooks = np.array(
    glob("../results/original/codebooks/*.npy")
), np.array(glob("../results/speckle/codebooks/*.npy"))

epochs = [int(file.split("_")[-1].split(".")[0]) for file in og_codebooks]
sort = np.argsort(epochs)
og_codebooks = og_codebooks[sort]

epochs = [int(file.split("_")[-1].split(".")[0]) for file in speckle_codebooks]
sort = np.argsort(epochs)
speckle_codebooks = speckle_codebooks[sort]
epochs = np.sort(epochs)


metrics = ["pca", "tsne", "mds", "isomap", "lle"]
save_paths = [f"../results/video/og_speckle/{metric}/" for metric in metrics]

for og, speckle, epoch in zip(og_codebooks, speckle_codebooks, epochs):
    _save_paths = [f"{path}{epoch}.png" for path in save_paths]
    if (
        os.path.exists(_save_paths[0])
        & os.path.exists(_save_paths[1])
        & os.path.exists(_save_paths[2])
        & os.path.exists(_save_paths[3])
        & os.path.exists(_save_paths[4])
    ):
        continue
    else:
        utils.visualize_codebooks_from_paths(
            [og, speckle],
            [f"Original", f"Speckle"],
            f"Epoch = {epoch}",
            _save_paths,
        )
```
