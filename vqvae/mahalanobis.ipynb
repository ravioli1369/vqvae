```python
import numpy as np
from scipy.spatial.distance import mahalanobis
import matplotlib.pyplot as plt

codebook_1 = np.load('../results/original/original.npy')
codebook_2 = np.load('../results/gaussian/gaussian.npy')

if codebook_1.size == 0 or codebook_2.size == 0:
    raise ValueError("One of the codebook files is empty.")

mean_vec = np.mean(codebook_1, axis=0)
cov_matrix = np.cov(codebook_1.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)

mahal_distances = [mahalanobis(v, mean_vec, inv_cov_matrix) for v in codebook_2]

plt.figure(figsize=(8, 6), dpi = 200)
plt.hist(mahal_distances, bins=30, color='black', alpha=0.7)

plt.title('Mahalanobis Distance between Clean and Noisy(Gaussian) Codebooks')
plt.xlabel('Mahalanobis Distance')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
from scipy.spatial.distance import mahalanobis
import matplotlib.pyplot as plt

codebook_1 = np.load('../results/original/original.npy')
codebook_2 = np.load('../results/poisson/poisson.npy')

if codebook_1.size == 0 or codebook_2.size == 0:
    raise ValueError("One of the codebook files is empty.")

mean_vec = np.mean(codebook_1, axis=0)
cov_matrix = np.cov(codebook_1.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)

mahal_distances = [mahalanobis(v, mean_vec, inv_cov_matrix) for v in codebook_2]

plt.figure(figsize=(8, 6), dpi = 200)
plt.hist(mahal_distances, bins=30, color='black', alpha=0.7)

plt.title('Mahalanobis Distance between Clean and Noisy(Poisson) Codebooks')
plt.xlabel('Mahalanobis Distance')
plt.ylabel('Frequency')
plt.show()
```

High Frequency Near Zero:

A large number of vectors (over 40) have Mahalanobis distances close to zero. This suggests that most of the noisy vectors are still very close to the clean distribution, meaning the noise hasn't significantly impacted these vectors.
Long Tail with Few Outliers:

There are a few vectors with larger Mahalanobis distances, reaching values up to 30. This indicates that, while the majority of the vectors remain close to the clean distribution, a small subset of vectors has been significantly affected by the noise. These outliers could represent vectors where noise has introduced considerable deviation.
Spread of Distances:

Beyond the main cluster near zero, there is a relatively spread-out distribution of distances, with a few vectors having moderate Mahalanobis distances (around 5 to 15). This could indicate that noise affects some vectors more than others, depending on their original characteristics.
Inference:
Most Vectors are Robust to Noise: Since the majority of the vectors have distances near zero, it suggests that Gaussian noise does not drastically affect most of the codebook vectors, and they remain similar to the clean data distribution.

Outliers and Noise Sensitivity: The vectors with larger Mahalanobis distances could be areas where the noise has a greater effect, potentially indicating a sensitivity to noise in certain regions of the codebook.

Potential Focus on Outliers: If you’re looking for which parts of the codebook are most affected by noise, it would be valuable to inspect the vectors corresponding to the higher Mahalanobis distances. These could represent cases where noise has caused a significant shift away from the clean distribution.
